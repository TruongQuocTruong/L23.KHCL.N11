```python
from sys import stdin
arr = list(stdin.readline().strip())
arr = [int(i) for i in arr]
arr.sort(reverse=False)
r=sum(arr)%3
if r!=0:
    dem=0
    l=len(arr)
    for i in range(0,l):
      if arr[i]%3==r:
        arr.remove(arr[i])
        dem+=5
        break
    if dem!=5:
        for i in range(0,l):
          if dem>3:
            dem+=5
            break
          else:
            for j in range(i+1,l):
              if arr[i] % 3 == 0: break
              if ((arr[i] + arr[j])%3==r):
                arr.remove(arr[i])
                arr.remove(arr[j-1])
                dem+=5
                break
arr.reverse()
print(*arr,sep="")
```
